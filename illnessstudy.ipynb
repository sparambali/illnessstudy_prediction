```python
#Load Data
illnessstudydata = pd.read_csv('./illnessstudy.csv')
illnessstudydata.head()
```

```python
#Create x and y variables
X = illnessstudydata.drop('diagnosis',axis=1).to_numpy()
y = illnessstudydata['diagnosis'].to_numpy()
```

```python
#Create Train and Test datasets
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y,test_size = 0.20,random_state=100)
```

```python
#Scale the data
from sklearn.preprocessing import StandardScaler  
sc = StandardScaler()  
x_train2 = sc.fit_transform(X_train)
x_test2 = sc.transform(X_test)
print("Done")
```

```python
#Script for SVM 
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix  

for name,method in [('SVM', SVC(kernel='linear',random_state=100))]: 
    method.fit(x_train2,y_train)
    predict = method.predict(x_test2)
    target_names=['M','B']
    print('\nEstimator: {}'.format(name)) 
    print(confusion_matrix(y_test,predict))  
    print(classification_report(y_test,predict,target_names=target_names)) 
```

```python
#Script for Logistic Regression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, plot_roc_curve  

name = 'Logistic Regression'
method = LogisticRegression(solver='liblinear',random_state=100)

method.fit(x_train2,y_train)
predict = method.predict(x_test2)
target_names=['Class 0','Class 1']
print(confusion_matrix(y_test,predict))  
print(classification_report(y_test,predict,target_names=target_names))
plot_roc_curve(method, x_test2, y_test)
```

```python
#Script for Decision Tree
from sklearn.tree import DecisionTreeClassifier  

for name,method in [('DT', DecisionTreeClassifier(random_state=100))]: 
    method.fit(x_train2,y_train)
    predict = method.predict(x_test2)
    target_names=['M','B']
    print('\nEstimator: {}'.format(name)) 
    print(confusion_matrix(y_test,predict))  
    print(classification_report(y_test,predict,target_names=target_names))
```
